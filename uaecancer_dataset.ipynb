## Cleaning Data

```python
# Drop the Death_Date column
dataset.drop('Death_Date',axis=1,inplace=True)
print(dataset)
```

```python
# Drop the Cause_of_Death column
dataset.drop('Cause_of_Death',axis=1,inplace=True)
print(dataset)
```

```python
# Delete the record which are blanks
dataset=dataset.dropna(how="all")
dataset
```

```python
# Delete the record which are blanks
dataset=dataset.dropna(how="any")
dataset
```

```python
# Rename the column Comorbidities to Medical_Conditions
dataset.rename(columns={'Comorbidities' :'Medical_Conditions'},inplace=True)
dataset
```

## Descriptive Analysis

```python
# Create a 20% sample
sample = dataset.sample(frac=0.2, random_state=42)
```

```python
sample['Patient_ID'].count() 
```

```python
# Weight mean
sample['Weight'].mean().round(1)
```

```python
# Weight median
sample['Weight'].median()
```

```python
# Weight mode
sample['Weight'].mode()
```

```python
# Weight standard deviation
sample['Weight'].std()
```

```python
# Weight variance
sample['Weight'].var()
```

```python
# Minimum Weight
sample['Weight'].min()
```

```python
# Maximum Weight
sample['Weight'].max()
```

```python
# find the different quantiles for the Weight
five_num = [
    sample["Weight"].quantile(0),    # Minimum (0th percentile)
    sample["Weight"].quantile(0.25), # Q1 (25th percentile)
    sample["Weight"].quantile(0.50), # Median (50th percentile)
    sample["Weight"].quantile(0.75), # Q3 (75th percentile)
    sample["Weight"].quantile(1)     # Maximum (100th percentile)
]
 
five_num 
```

```python
# find the inter quartile range(IQR) for Weight
sample["Weight"].quantile(0.75) - sample["Weight"].quantile(0.25) 
```

```python
# Weight skewness
sample['Weight'].skew()
```

```python
# Weight kurtosis
sample['Weight'].kurt()
```

## Box Plot

```python
# Draw the box plot for quartiles of Weight using mat plot library
import matplotlib.pyplot as plt 
sample.boxplot(column="Weight", 
               return_type='axes', 
               figsize=(8,8)) 

plt.savefig('box plot for quartiles of Weight.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sample['Cancer_Type'].value_counts().plot(kind='pie', title='Cancer_Type')
plt.savefig('Cancer_Type.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Distribution plot using histplot with KDE
plt.figure(figsize=(8, 6))
sns.histplot(data=sample, x='Age', kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.tight_layout()
plt.savefig('#histplot with KDE.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a single scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Weight', data=sample)
plt.title('Age vs. Weight')
plt.xlabel('Age')
plt.ylabel('Weight')

plt.tight_layout()
plt.savefig('Age_vs_Weight_Scatter.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a single boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Outcome', y='Age', data=sample)
plt.title('Age by Outcome Status')
plt.xlabel('Outcome')
plt.ylabel('Age')

plt.tight_layout()
plt.savefig('Age_by_Outcome_Boxplot.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("uaecancer_dataset.csv")

# Take 20% sample
sample_df = df.sample(frac=0.2, random_state=42)

# Select only numeric columns (exclude Patient IDs, text, etc.)
numeric_df = sample_df.select_dtypes(include=['float64', 'int64'])

# Generate correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (20% Sample)")
plt.savefig('Correlation Heatmap (20% Sample).png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.countplot(x='Cancer_Type', data=sample)
plt.title("Cancer Type Distribution (20% Sample)")
plt.xlabel("Cancer Type")
plt.ylabel("Count")
plt.xticks(rotation=45)  
plt.show()
```
